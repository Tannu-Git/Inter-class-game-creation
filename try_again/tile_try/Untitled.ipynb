```python
import csv
with open('level_1s.csv' , mode= 'r') as l:
    reader = csv.reader(l)
    data = list(reader)
for i in range(len(data)):
#     print(data[i])
    for o in range(len(data[i])) :
        if data[i][o] == "0":
#             print(data[i][o])
            data[i][o] = "c"
        if data[i][o] == "36":
            data[i][o] = "y" 
        if data[i][o] == "8":
            data[i][o] = "e" 
            
#  change value 0 to c - coins
print(data)
```
